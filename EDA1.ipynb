```python
# 2. Statistical Summary:
#◦ Provide a statistical summary for each variable in the dataset, including measures of central tendency 
   # (mean, median) and dispersion (standard deviation, interquartile range).
#◦ Highlight any interesting findings from this summary.
```

```python
for c in data.columns:
  column_name = c
  print('Outliers calculation for', c)
  print(data[c].describe())
#Quartile 1
  Q1 = data[c].quantile(0.25)
  print('Quartile 1 : ',Q1)
#Quartile 3
  Q3 = data[c].quantile(0.75)
  print('Quartile 3 : ',Q3)
#Interquartile Range
  IQR = Q3 - Q1
  print('Inter-Quartile Range : ',IQR)
#Upper Limit
  UL = Q3 + (1.5 * IQR)
  print('Upper Limit : ',UL)
#Lower Limit
  LL = Q1 - (1.5 * IQR)
  print('Lower Limit : ', LL)
  Outliers = data[(data[c] > UL) | (data[c] < LL)]
  print('The % Outliers for column', c, 'are:', round(((len(Outliers) / len(data[c])) * 100), 4))
  print('__________________________________________________________ \n__________________________________________________________')
```

```python
# To check outliers in other variables
for c in data.columns:
    sea.boxplot(x = data[c])
    plt.show()
```

```python
# Maximum outliers are in NSP, FM, ALTV, and DP data sets.
# It is interesting to note that the data in NSP, FM, and DP columns are appears to be narrowly spread meaning 
# the inter-quartile range is narrow and this may be a reason for smaller upper limit and hence more outliers. 
# In the case of ALTV data, the data appears to be skewed towards the lower limit which creates an imbalance 
# and results in a larger upper limit along with outliers. 
# More insights might be obtained with histagrams and density distributions.
```

```python
# To check for statistical summary for each column
data.describe()
```

```python
# To show histograms and normalized 
for c in data.columns:
    ax = sea.histplot(data = data, x = c, kde = True, stat = 'probability', alpha = 0.6)
    ax.set_xlim((data[c].min()), (data[c].max() * 1.1)) 
    plt.show()

# Zoomed FM plot    
ax = sea.histplot(data = data, x = data['FM'], kde = True, stat = 'probability', alpha = 0.6)
ax.set_ylim(0, 0.012)
plt.title('Zoomed plot of FM')
plt.show()
```

```python
# To show correlation heatmaps
corr_matrix = data.corr()
sea.heatmap(corr_matrix, annot=True, cmap='icefire', fmt='.1f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sea.pairplot(data, kind = 'reg', diag_kind='kde')  # kde for density plot
plt.show()
```

```python
# Based on the correlation heatmap, we can see the highest correation between two variables is +0.4, 
# which is generally considerd as moderate positive correlation
# and -0.3 indicates weak negetive correlation.
# To explore more about them I am using scatter plots with linear regression line to show the trend.
```
