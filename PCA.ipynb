**Task 2: Dimensionality Reduction with PCA:**

* Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
* Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
* Transform the original dataset into the principal components.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
```

```python
x_data = wine_data.drop('Type', axis=1)
y_data = wine_data['Type']
```

Standardize the data set after segregating 'dependent' variable.

```python
std_scaler = StandardScaler()
x_data_scaler = std_scaler.fit_transform(x_data)
```

```python
# Identifying optimal number of principal components using Scree plot (Elbow method)
pca = PCA()
pca_data = pca.fit(x_data_scaler)
plt.plot(pca_data.explained_variance_, 'o-')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance')
plt.title('Scree Plot')
plt.show()
```

Based on Scree plot (Elbow method), 3 to 4 PCs are required to capture most of the variance in the data.

```python
plt.plot(np.cumsum(pca_data.explained_variance_ratio_), 'o-')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y = 0.90, color = 'r', linestyle = '--')
plt.title('Cumulative Explained Variance')
plt.show()
```

This plots shows that 6 PCs are required to capture about 90% of the variance.

```python
# Transforming original data into principal components
pca_01 = PCA(n_components = 3)
pca_data_01 = pca_01.fit_transform(x_data_scaler)
print(pca_data_01[:5])

pca_02 = PCA(n_components = 6)
pca_data_02 = pca_02.fit_transform(x_data_scaler)
print('\n', pca_data_02[:5])
```

**Task 3: Clustering with Original Data:**

* Apply a clustering algorithm (e.g., K-means) to the original dataset.
* Visualize the clustering results using appropriate plots.
* Evaluate the clustering performance using metrics such as silhouette score or Daviesâ€“Bouldin index.

```python
# Calculating Within Cluster Sum-of-Squares
wcss = []

for i in range(1, 21):
  kmeans = KMeans(n_clusters = i, random_state = 0)
  kmeans.fit(x_data_scaler)
  wcss.append(kmeans.inertia_)

plt.plot(range(1, 21), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

To have an idea of how many clusters would be required to apply K-means clustering WCSS vs Number of clusters plot is plotted. The 1st elbow appears at 3 (number of clusters).

Building K-means model

```python
model_01 = KMeans(n_clusters = 3, random_state = 0)
model_01.fit(x_data_scaler)
```

```python
wine_data['Cluster_m01'] = model_01.labels_
wine_data
```

```python
# To get the count of number of data points in each cluster
wine_data['Cluster_m01'].value_counts()
```

```python
# To get the mean of all variables (columns) cluster wise
wine_data.groupby('Cluster_m01').mean()
```

The clustering was performed using the data 'dependent' seperated data. By getting the mean of the 'dependent' variable 'Type', evaluate the clustering.  

```python
# To get number of data points with each Type in the original data
count_1, count_2, count_3 = 0, 0, 0
for i in range(len(wine_data)):
  if wine_data['Type'][i] == 1:
    count_1 += 1
  elif wine_data['Type'][i] == 2:
    count_2 += 1
  elif wine_data['Type'][i] == 3:
    count_3 += 1

print('Number of data points with Type-1:', count_1, ', Type-2:', count_2, ', Type-3:', count_3)
```

Based on the obtained results cluster-0 has Type mean value of 2, but total number of Type: 2 are 71 and cluster-0 are 65, meaning 6 data points are not clustered in different groups.

```python
# Check if there are any other Type (1 or 3) in cluster-0
for i in range(len(wine_data)):
  if wine_data['Cluster_m01'][i] == 0 and wine_data['Type'][i] != 2:
    print(i, wine_data['Type'][i])
```

```python
# Check if there are any other Type (2 or 1) in cluster-1
for i in range(len(wine_data)):
  if wine_data['Cluster_m01'][i] == 1 and wine_data['Type'][i] != 3:
    print(i, wine_data['Type'][i])
```

```python
# Check if there are any other Type (3 or 1) in cluster-2
for i in range(len(wine_data)):
  if wine_data['Cluster_m01'][i] == 2 and wine_data['Type'][i] != 1:
    print(i, wine_data['Type'][i])
```

```python
plt.scatter(wine_data['Type'], wine_data['Cluster_m01'], c = model_01.labels_)
plt.xlabel('Type')
plt.ylabel('Cluster')
plt.show()
```

```python
plt.scatter(wine_data['Type'], wine_data['Phenols'], c = model_01.labels_)
plt.xlabel('Type')
plt.ylabel('Phenols')
plt.show()
```

The missing 6 data with Type-2 are accounted, which means about 91.5% of the data are clustered accuratly (based on Type variable).

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
sil_score_01 = silhouette_score(x_data_scaler, model_01.labels_)
dav_score_01 = davies_bouldin_score(x_data_scaler, model_01.labels_)

print('Silhouette Score:', round(sil_score_01, 4))
print('Davies-Bouldin Score:', round(dav_score_01, 4))
```

The Silhouette score measures the average distance between data points in the same cluster and average distance to datapoints in different cluster, which represent cohesion and seperation. Higher the score the better is the model, this K-means model is in the fair range and so room for improvement. Additionally, more than 90% of the data is assigned into approriate clusters.
